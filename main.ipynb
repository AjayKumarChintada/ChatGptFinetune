```python
import pandas as pd 
```

```python
df = pd.read_csv('Mental_Health_FAQ.csv')
df
```

```python
df = df[['Questions','Answers']]
df.head(10)
```

```python
import unicodedata

def remove_unwanted_data(str):
    return str.replace("\xa0", " ")


def fix_unicode(string):
    return unicodedata.normalize("NFKD", string).encode("ascii", "ignore").decode("utf-8")
```

```python
df = df.apply(lambda row : remove_unwanted_data(row))
```

```python
df
```

```python
data  = []
for _, row in df.iterrows():
    completion = fix_unicode( row['Answers'])
    completion = completion.split('\n')
    completion = ' '.join(completion)
    convo = {'prompt': fix_unicode( row['Questions']),'completion':completion}
    data.append(convo)
    
```

```python
import json
with open('mental_health.jsonl', 'w') as outfile:
        for i in data:
            json.dump(i, outfile)
            outfile.write('\n')
```
